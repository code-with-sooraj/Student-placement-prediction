# Importing the necessary modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

# Reading the raw dataset 

```python
dataset = pd.read_csv('./Placement_Data_Full_Class.csv')
```

```python
dataset
```

```python
temp=dataset
```

# Removing irrelevant cols

```python
dataset.drop(['sl_no','salary'],axis=1,inplace=True)
```

```python
dataset
```

# Data preprocessing

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

# Filling the missing values

```python
mean_ssc_p = dataset['ssc_p'].mean()
print('Mean of ssc_p: {}'.format(mean_ssc_p))

mean_hsc_p = dataset['hsc_p'].mean()
print('Mean of hsc_p: {}'.format(mean_hsc_p))

mode_hsc_s = dataset['hsc_s'].mode()[0]
print('Mode of hsc_s: {}'.format(mode_hsc_s))

mean_degree_p = dataset['degree_p'].mean()
print('Mean of degree_p: {}'.format(mean_degree_p))

mode_degree_t = dataset['degree_t'].mode()[0]
print('Mode of degree_t: {}'.format(mode_degree_t))

mode_workex = dataset['workex'].mode()[0]
print('Mode of workex: {}'.format(mode_workex))

mean_etest_p = dataset['etest_p'].mean()
print('Mean of etest_p: {}'.format(mean_etest_p))

mode_status = dataset['status'].mode()[0]
print('Mode of status: {}'.format(mode_status))
```

```python
dataset['ssc_p'].fillna(mean_ssc_p,inplace=True)
dataset['hsc_p'].fillna(mean_hsc_p,inplace=True)
dataset['hsc_s'].fillna(mode_hsc_s,inplace=True)
dataset['degree_p'].fillna(mean_degree_p,inplace=True)
dataset['degree_t'].fillna(mode_degree_t,inplace=True)
dataset['workex'].fillna(mode_workex,inplace=True)
dataset['etest_p'].fillna(mean_etest_p,inplace=True)
dataset['status'].fillna(mode_status,inplace=True)
```

# converts categorical data into dummy or indicator variables.


```python
features_to_split = ['hsc_s','degree_t']
for feature in features_to_split:
    dummy = pd.get_dummies(dataset[feature],drop_first=True)
    dataset = pd.concat([dataset, dummy], axis=1)
    dataset.drop(feature, axis=1, inplace=True)
```

```python
dataset
```

```python
dataset.rename(columns={"Others": "Other_Degree"},inplace=True)
```

```python
dataset
```

# label encoding needs to be done to ensure all  categorical values in the dataset is numeric


```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label =LabelEncoder()
```

```python
cols_to_encode = ['gender','ssc_b','hsc_b','workex','specialisation','status']
for col in cols_to_encode:
    dataset[col]=label.fit_transform(dataset[col])
```

```python
dataset
```

```python
dataset.describe()
```

# Checking for outliers

```python
fig, axs = plt.subplots(ncols=5,nrows=3,figsize=(20,10))
index = 0
axs = axs.flatten()
for k,v in dataset.items():
    sns.boxplot(y=v, ax=axs[index])
    index+=1

plt.tight_layout(pad=0.3, w_pad=0.5,h_pad = 4.5)
```

```python
dataset = dataset[~(dataset['degree_p']>=90)]
dataset = dataset[~(dataset['hsc_p']>=95)]
```

# Checking for correlation

# Visual representation of correlation

```python
x = dataset.loc[:,dataset.columns!='status'] 
y = dataset.loc[:, 'status']
```

```python
x
```

```python
y
```

# Standardization of all features

```python
sc= StandardScaler()
x_scaled = sc.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.20,random_state=42)
```

# Using Logistic Regression


```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
accuracy_score(y_test,y_pred)
```

```python
lr.score(x_train,y_train)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# Using Naive Bayes Classifier - Gaussian Naive Bayes

```python
nbclassifier = GaussianNB()
```

```python
nbclassifier.fit(x_train, y_train)
```

```python
y_pred_nb = nbclassifier.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_nb)
```

```python
nbclassifier.score(x_train, y_train)
```

```python
confusion_matrix(y_test, y_pred_nb)
```

```python
print(classification_report(y_test,y_pred_nb))
```

# Using SVM Linear Kernel

```python
clf = svm.SVC(kernel="linear")
clf.fit(x_train, y_train)
```

```python
y_pred_svm = clf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred_svm)
```

```python
clf.score(x_train, y_train)
```

```python
confusion_matrix(y_test, y_pred_svm)
```

```python
print(classification_report(y_test, y_pred_svm))
```
